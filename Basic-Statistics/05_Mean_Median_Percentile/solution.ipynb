# Solution

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Set random seed for reproducibility
np.random.seed(42)
```

## Problem Analysis

<!-- Add your analysis here -->

When would you use the mean, median and percentile?

```python
# Generate synthetic skewed data (log-normal distribution)
data = np.random.lognormal(mean=3.0, sigma=1.0, size=1000)

# Compute summary statistics
mean_val = np.mean(data)
median_val = np.median(data)
p10 = np.percentile(data, 10)
p90 = np.percentile(data, 90)
p95 = np.percentile(data, 95)
p99 = np.percentile(data, 99)

# Display results
print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"10th Percentile: {p10:.2f}")
print(f"90th Percentile: {p90:.2f}")
print(f"95th Percentile: {p95:.2f}")
print(f"99th Percentile: {p99:.2f}")
```

## Visualization

<!-- Add visualizations if applicable -->

```python
plt.figure(figsize=(10, 6))
plt.hist(data, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Median: {median_val:.2f}')
plt.axvline(p10, color='orange', linestyle=':', linewidth=1.5, label=f'p10: {p10:.2f}')
plt.axvline(p90, color='purple', linestyle=':', linewidth=1.5, label=f'p90: {p90:.2f}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Skewed Distribution: Mean vs Median vs Percentiles')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

## Conclusion

<!-- Summarize findings and insights -->

Normal distribution?
- Use mean or median (they'll be nearly identical)
Skewed distribution or outliers present?
- Use median for central tendency
- Report mean alongside for completeness, but don't rely on it alone
Need tail insights (low/high extremes)?
- Use percentiles (p10, p90, p95, p99, etc.)
